## Introduction to Jupyter Notebooks

**What is a Jupyter Notebook?**

It is a tool that lets you **write text** and **run computer code** (typically R or Python) all mixed into a single file, making it easy to see and understand. That is, Instead of just writing programming code in one tool and the descriptions or explanations in another file or a textbook, it blends them together sequentially.
<br>
Also, Jupyter lets you type that computing code into boxes (called "cells") and run each part separately. 
<br>
As it turns out, **this page** was written as a Jupyter Notebook but as the user, you only see it as a web page. Since it doesn't have any code (yet) to run, it basically *is* a web page.
<br>
BUT, when you look at a Jupyter notebook in the right kind of interface, the user can edit the text not just view it.
<br>
<i>**and that user can change the code and run it!**</i>
<br)
That is, With Jupyter Notebook, you can add notes, explanations, and even pictures to help make sense of your work. This makes it a great tool for learning, experimenting, and sharing code with others.

Think of it like a notebook where you can write instructions (code), run them step by step, and take notesâ€”all in one place!
